```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("County_Health_Rankings.csv")
df
```

```python
df.fillna(0,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

### 1)find out top 5 highest health Ranking by state

```python
high_rank_by_state =df.groupby('State')['Raw value'].max().sort_values(ascending=False)

top_5_states = high_rank_by_state.head(5)
print(top_5_states)

colors = ['lightgreen', 'salmon', 'gold', 'lightcoral', 'lightblue',]
plt.figure(figsize=(10, 6))
top_5_states.plot(kind='bar', color=colors)
plt.title('Top 5 highest Health Rankings by State')
plt.xlabel('State')
plt.ylabel('Health Ranking')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 2)What is average raw value for each year span?

```python
avg_raw_value=df.groupby('Year span')['Raw value'].mean()
avg1=avg_raw_value.sort_values(ascending=False).head(3)
print("Average Raw value of a specific measures across all ountries and years by",avg1)

plt.figure(figsize=(12,6))
plt.bar(avg1.index,avg1.values,color=['lightgreen','red','pink','purple'],width=0.8,label=avg1.index)
plt.xlabel('Year span')
plt.ylabel('Raw Value')
plt.xticks()
plt.show()
```

### 3)Which state has the lowest average health ranking?

```python
avg_health_ranking_by_state = df.groupby('State')['Raw value'].mean().sort_values(ascending=False)
lowest_avg_health_ranking_state = avg_health_ranking_by_state.idxmin()
print("State with the lowest average health ranking:", lowest_avg_health_ranking_state)
```

### 4)find out top 5 highest health Ranking by country

```python
high_rank =df.groupby('County')['Raw value'].max().sort_values(ascending=False)

top_5_states = high_rank.head(5)
print(top_5_states)

colors = ['lightgreen', 'salmon', 'gold', 'lightcoral', 'lightblue',]
plt.figure(figsize=(10, 6))
top_5_states.plot(kind='bar', color=colors)
plt.title('Top 5 highest Health Rankings by County')
plt.xlabel('County')
plt.ylabel('Health Ranking')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 5)find out top 5 highest confidence interval range for specific country?

```python
max_confidence=df.groupby('County')['Confidence Interval Upper Bound'].max()
avg1=max_confidence.sort_values(ascending=False).head()
print("Top 5 Confidence interval range by",avg1)

plt.figure(figsize=(10,6))
plt.bar(avg1.index,avg1.values,color=['lightgreen','red','pink','salmon', 'gold'],width=0.8,label=avg1.index)
plt.title("Top 5 Confidence interval range by countries")
plt.xlabel('County')
plt.ylabel('Confidence Interval Upper Bound')
plt.xticks()
plt.show()
```

### 6)What is the correlation between health rankings and the numerator/denominator?


```python

correlation = df[['Raw value', 'Numerator', 'Denominator']].corr()
print("Correlation between health rankings and numerator/denominator:")
print(correlation)
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation between Health Rankings and Numerator/Denominator')
plt.show()
```

### 7)CalCulate maximum count of top 5 states with specific confidence interval range?

```python
max_confidence=df.groupby('State')['Confidence Interval Lower Bound'].max()
avg1=max_confidence.sort_values(ascending=False).head(5)
print("Top 5 Confidence interval range by",avg1)

plt.figure(figsize=(10,6))
plt.bar(avg1.index,avg1.values,color=['lightgreen','red','pink','purple','brown'],width=0.8,label=avg1.index)
plt.title("Top 3 Confidence interval range by States")
plt.xlabel('State')
plt.ylabel('Confidence Interval Lower Bound')
plt.xticks()
plt.show()
```

### 8)Calculate the maximum count of top 2 countries with data release year?

```python
avg_data_release=df.groupby('County')['Data Release Year'].max()
avg1=avg_data_release.sort_values(ascending=False).head(2)
print("Top 2 data release year for each ",avg1)

plt.figure(figsize=(10,6))
plt.bar(avg1.index,avg1.values,color=['lightgreen','red'],width=0.8,label=avg1.index)
plt.title("Top 2 Data release year for each country")
plt.xlabel('County')
plt.ylabel('Data Release Year')
plt.xticks()
plt.show()
```

### 9)Calculate maximum count of bottom 5 fipscode with sepcific country name?

```python
max_fipscode=df.groupby('County')['fipscode'].max()
avg1=max_fipscode.sort_values(ascending=False).tail(5)
print("Bottom 5 Fipscode with the help of ",avg1)

plt.figure(figsize=(10,6))
plt.bar(avg1.index,avg1.values,color=['lightgreen','red','pink','brown'],width=0.8,label=avg1.index)
plt.title("Bopttom 5 fipscode with the help of country name")
plt.xlabel('County')
plt.ylabel('fipscode')
plt.xticks()
plt.show()
```
